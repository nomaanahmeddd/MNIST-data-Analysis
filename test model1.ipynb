```python
from keras.models import load_model
```

```python
trained_model = load_model('digit_recog-1epch.h5')
```

```python
trained_model
```

```python
image = 'MNIST - JPG - testing/7/9990.jpg'
```

```python
import cv2
```

```python
from matplotlib import pyplot as plt
img = cv2.imread(image,cv2.COLOR_BGR2GRAY)
IMAGE_SIZE = 256
img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
plt.figure()
plt.axis('Off')
plt.imshow(img)
```

```python
import numpy as np
img = np.array(img) / 255.
img = img.reshape((1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

```python
from keras.preprocessing.image import ImageDataGenerator
data_generator = ImageDataGenerator(
                samplewise_center=True, 
                samplewise_std_normalization=True,
                brightness_range=[0.8, 1.0],
                zoom_range=[1.0, 1.2],
                validation_split=0.1
            )
```

```python
img = data_generator.standardize(img)
```

```python
result = trained_model.predict(img)
```
